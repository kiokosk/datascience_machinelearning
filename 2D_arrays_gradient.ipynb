```python
import numpy as np
import matplotlib.pyplot as plt

# Problem 1: Linear functions
def problem1():
    # Create x from -50 to 50 with 0.1 intervals
    x = np.arange(-50, 50.1, 0.1)
    
    # Calculate y = 12x + 1
    y = 12 * x + 1
    
    return x, y
```

```python
# Problem 2: Array concatenation
def problem2(x, y):
    # Combine x and y into a single array of shape (1001, 2)
    xy = np.column_stack((x, y))
    
    return xy
```

```python
# Problem 3: Find the gradient
def problem3(x, y):
    # Calculate the gradient as the change in y divided by the change in x
    gradient = np.diff(y) / np.diff(x)
    
    return gradient
```

```python
# Problem 4: Draw graphs
def problem4(x, y, gradient):
    # Create figure with two subplots
    fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(15, 6))
    
    # Plot the function y = 12x + 1
    ax1.plot(x, y)
    ax1.set_title("Linear Function: y = 12x + 1")
    ax1.set_xlabel("x")
    ax1.set_ylabel("y")
    ax1.grid(True)
    
    # Plot the gradient 
    # Note: gradient has one less point than x, so we use x[:-1] or x[1:]
    # Using x[:-1] means we associate gradient with the start of each interval
    ax2.plot(x[:-1], gradient)
    ax2.set_title("Gradient of y = 12x + 1")
    ax2.set_xlabel("x")
    ax2.set_ylabel("Gradient")
    ax2.grid(True)
    
    plt.tight_layout()
    plt.savefig('linear_function_gradient.png')
    plt.show()
```

```python
# Problem 5: Python function to compute gradient
def compute_gradient(function, x_range=(-50, 50.1, 0.1)):
    """
    Calculate the gradient using the rate of change.
    
    Parameters
    ----------------
    function : function
        The function to compute, which takes an ndarray of x and returns an ndarray of y.
    x_range : tuple
        Range specification in the same format as np.arange() arguments.
        
    Returns
    ----------------
    array_xy : ndarray, shape(n, 2)
        Combined array of x and y values. n depends on x_range.
    gradient : ndarray, shape(n-1,)
        Gradient of the function. Has one fewer element due to differencing.
    """
    # Create x array based on range
    x = np.arange(*x_range)
    
    # Compute y values
    y = function(x)
    
    # Combine x and y
    array_xy = np.column_stack((x, y))
    
    # Calculate gradient
    gradient = np.diff(y) / np.diff(x)
    
    return array_xy, gradient

# Define the functions from Problem 5
def function1(array_x):
    """y = x^2"""
    return array_x ** 2

def function2(array_x):
    """y = 2x^2 + 2x"""
    return 2 * array_x ** 2 + 2 * array_x

def function3(array_x):
    """y = sin(x/12)"""
    return np.sin(array_x / 12)
```

```python
def problem6():
    # Function 1: y = x^2
    array_xy1, gradient1 = compute_gradient(function1)
    min_val1 = array_xy1[:, 1].min()
    min_idx1 = array_xy1[:, 1].argmin()
    min_x1 = array_xy1[min_idx1, 0]
    
    # Get gradients before and after the minimum
    if min_idx1 > 0 and min_idx1 < len(gradient1):
        grad_before1 = gradient1[min_idx1-1]
        grad_after1 = gradient1[min_idx1]
    else:
        grad_before1 = None
        grad_after1 = gradient1[0] if min_idx1 == 0 else None
    
    # Function 2: y = 2x^2 + 2x
    array_xy2, gradient2 = compute_gradient(function2)
    min_val2 = array_xy2[:, 1].min()
    min_idx2 = array_xy2[:, 1].argmin()
    min_x2 = array_xy2[min_idx2, 0]
    
    # Get gradients before and after the minimum
    if min_idx2 > 0 and min_idx2 < len(gradient2):
        grad_before2 = gradient2[min_idx2-1]
        grad_after2 = gradient2[min_idx2]
    else:
        grad_before2 = None
        grad_after2 = gradient2[0] if min_idx2 == 0 else None
    
    # Function 3: y = sin(x/12)
    array_xy3, gradient3 = compute_gradient(function3, (0, 50.1, 0.1))
    min_val3 = array_xy3[:, 1].min()
    min_idx3 = array_xy3[:, 1].argmin()
    min_x3 = array_xy3[min_idx3, 0]
    
    # Get gradients before and after the minimum
    if min_idx3 > 0 and min_idx3 < len(gradient3):
        grad_before3 = gradient3[min_idx3-1]
        grad_after3 = gradient3[min_idx3]
    else:
        grad_before3 = None
        grad_after3 = gradient3[0] if min_idx3 == 0 else None
    
    return {
        'function1': {
            'min_x': min_x1,
            'min_val': min_val1,
            'grad_before': grad_before1,
            'grad_after': grad_after1
        },
        'function2': {
            'min_x': min_x2,
            'min_val': min_val2,
            'grad_before': grad_before2,
            'grad_after': grad_after2
        },
        'function3': {
            'min_x': min_x3,
            'min_val': min_val3,
            'grad_before': grad_before3,
            'grad_after': grad_after3
        }
    }
```

```python
# Problems 1-4: Linear function gradient
x, y = problem1()
xy = problem2(x, y)
gradient = problem3(x, y)
problem4(x, y, gradient)

# Problem 5: Compute gradients for the three functions
print("\nProblem 5: Gradients for the three functions")

# Function 1: y = x^2
array_xy1, gradient1 = compute_gradient(function1)
print(f"\nFunction 1 (y = x^2):")
print(f"Shape of array_xy: {array_xy1.shape}")
print(f"Shape of gradient: {gradient1.shape}")
print(f"First few gradient values: {gradient1[:5]}")
print(f"Middle gradient values around x=0: {gradient1[499:502]}")
print(f"Last few gradient values: {gradient1[-5:]}")

# Function 2: y = 2x^2 + 2x
array_xy2, gradient2 = compute_gradient(function2)
print(f"\nFunction 2 (y = 2x^2 + 2x):")
print(f"Shape of array_xy: {array_xy2.shape}")
print(f"Shape of gradient: {gradient2.shape}")
print(f"First few gradient values: {gradient2[:5]}")
print(f"Middle gradient values around x=0: {gradient2[499:502]}")
print(f"Last few gradient values: {gradient2[-5:]}")

# Function 3: y = sin(x/12)
array_xy3, gradient3 = compute_gradient(function3, (0, 50.1, 0.1))
print(f"\nFunction 3 (y = sin(x/12)):")
print(f"Shape of array_xy: {array_xy3.shape}")
print(f"Shape of gradient: {gradient3.shape}")
print(f"First few gradient values: {gradient3[:5]}")
print(f"Last few gradient values: {gradient3[-5:]}")

# Plot graphs for all three functions
def plot_all_functions():
    fig, axes = plt.subplots(3, 2, figsize=(15, 15))
    
    # Function 1: y = x^2
    axes[0, 0].plot(array_xy1[:, 0], array_xy1[:, 1])
    axes[0, 0].set_title("Function 1: y = x^2")
    axes[0, 0].set_xlabel("x")
    axes[0, 0].set_ylabel("y")
    axes[0, 0].grid(True)
    
    axes[0, 1].plot(array_xy1[:-1, 0], gradient1)
    axes[0, 1].set_title("Gradient of Function 1")
    axes[0, 1].set_xlabel("x")
    axes[0, 1].set_ylabel("Gradient")
    axes[0, 1].grid(True)
    
    # Function 2: y = 2x^2 + 2x
    axes[1, 0].plot(array_xy2[:, 0], array_xy2[:, 1])
    axes[1, 0].set_title("Function 2: y = 2x^2 + 2x")
    axes[1, 0].set_xlabel("x")
    axes[1, 0].set_ylabel("y")
    axes[1, 0].grid(True)
    
    axes[1, 1].plot(array_xy2[:-1, 0], gradient2)
    axes[1, 1].set_title("Gradient of Function 2")
    axes[1, 1].set_xlabel("x")
    axes[1, 1].set_ylabel("Gradient")
    axes[1, 1].grid(True)
    
    # Function 3: y = sin(x/12)
    axes[2, 0].plot(array_xy3[:, 0], array_xy3[:, 1])
    axes[2, 0].set_title("Function 3: y = sin(x/12)")
    axes[2, 0].set_xlabel("x")
    axes[2, 0].set_ylabel("y")
    axes[2, 0].grid(True)
    
    axes[2, 1].plot(array_xy3[:-1, 0], gradient3)
    axes[2, 1].set_title("Gradient of Function 3")
    axes[2, 1].set_xlabel("x")
    axes[2, 1].set_ylabel("Gradient")
    axes[2, 1].grid(True)
    
    plt.tight_layout()
    plt.savefig('all_functions_gradients.png')
    plt.show()

plot_all_functions()

# Problem 6: Find minimum values
min_results = problem6()

print("\nProblem 6: Minimum values for the three functions")
for func_name, results in min_results.items():
    print(f"\n{func_name}:")
    print(f"Minimum value occurs at x = {results['min_x']}")
    print(f"Minimum value = {results['min_val']}")
    print(f"Gradient just before minimum: {results['grad_before']}")
    print(f"Gradient just after minimum: {results['grad_after']}")
```
