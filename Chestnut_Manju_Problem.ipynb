```python
import math
import matplotlib.pyplot as plt

def time_to_fill_space(initial_volume, target_volume, doubling_time=5):
    """
    Calculate the time required for an object's volume to exceed a given space.

    Parameters:
    - initial_volume (float): Initial volume of the object in cubic meters.
    - target_volume (float): Volume of the space to be filled in cubic meters.
    - doubling_time (int): Time (minutes) it takes for the object to double in volume.

    Returns:
    - total_minutes (int): Total time in minutes to fill the space.
    - steps (list): Volume values at each time step (for visualization).
    """
    n = 0
    current_volume = initial_volume
    volumes = [current_volume]

    while current_volume < target_volume:
        current_volume *= 2
        volumes.append(current_volume)
        n += 1

    total_minutes = n * doubling_time
    return total_minutes, volumes

def plot_growth(volumes, doubling_time):
    """
    Plot the exponential growth of volume over time.

    Parameters:
    - volumes (list): List of volume values over time.
    - doubling_time (int): Time interval for each doubling in minutes.
    """
    time_steps = [i * doubling_time for i in range(len(volumes))]
    
    plt.figure(figsize=(10, 5))
    plt.plot(time_steps, volumes, marker='o', linestyle='-', color='b', label="Volume Growth")
    plt.yscale("log")  
    plt.xlabel("Time (minutes)")
    plt.ylabel("Volume (m続)")
    plt.title("Exponential Growth of Chestnut Bun Volume")
    plt.legend()
    plt.grid(True)
    plt.show()

# Define volumes in cubic meters
V_CHESTNUT_BUN = 0.0005  # Example: 0.5 liters = 0.0005 m続
V_SOLAR_SYSTEM = 1.2e39  # Approximate volume of the solar system in m続
V_TOKYO_DOME = 1.24e6  # Volume of Tokyo Dome in m続

# Calculate time for chestnut bun to fill the solar system
time_solar, volumes_solar = time_to_fill_space(V_CHESTNUT_BUN, V_SOLAR_SYSTEM)
print(f"Time to fill the Solar System: {time_solar} minutes ({time_solar/60:.2f} hours)")

# Calculate time for chestnut bun to fill Tokyo Dome
time_dome, volumes_dome = time_to_fill_space(V_CHESTNUT_BUN, V_TOKYO_DOME)
print(f"Time to fill Tokyo Dome: {time_dome} minutes ({time_dome/60:.2f} hours)")

# Plot results
plot_growth(volumes_dome, 5)
```
