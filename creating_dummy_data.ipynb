```python
import numpy as np
import matplotlib.pyplot as plt

# Problem 1: Creating random numbers with a two-dimensional normal distribution
np.random.seed(0)  
mean1 = [-3, 0]  # Mean for first dataset
covariance = [[1.0, 0.8], [0.8, 1.0]]  # Covariance matrix

data1 = np.random.multivariate_normal(mean1, covariance, 500)
```

```python
# Problem 2: Scatter plot visualization
plt.figure(figsize=(8, 6))
plt.scatter(data1[:, 0], data1[:, 1], alpha=0.5, label='Dataset 1')
plt.xlabel("X-axis")
plt.ylabel("Y-axis")
plt.title("Scatter Plot of First Dataset")
plt.legend()
plt.show()
```

```python
# Problem 3: Histogram visualization
plt.figure(figsize=(12, 5))
plt.subplot(1, 2, 1)
plt.hist(data1[:, 0], bins=30, alpha=0.7)
plt.xlabel("X-axis")
plt.ylabel("Frequency")
plt.title("Histogram of X values")
plt.xlim(-6, 3)

plt.subplot(1, 2, 2)
plt.hist(data1[:, 1], bins=30, alpha=0.7)
plt.xlabel("Y-axis")
plt.ylabel("Frequency")
plt.title("Histogram of Y values")
plt.xlim(-3, 3)
plt.show()
```

```python
# Problem 4: Adding another dataset
mean2 = [0, -3]  # Mean for second dataset
data2 = np.random.multivariate_normal(mean2, covariance, 500)

# Visualizing both datasets in a single scatter plot
plt.figure(figsize=(8, 6))
plt.scatter(data1[:, 0], data1[:, 1], alpha=0.5, label='Dataset 1')
plt.scatter(data2[:, 0], data2[:, 1], alpha=0.5, label='Dataset 2')
plt.xlabel("X-axis")
plt.ylabel("Y-axis")
plt.title("Scatter Plot of Both Datasets")
plt.legend()
plt.show()
```

```python
combined_data = np.vstack((data1, data2))  
```

```python
# Problem 6: Labeling
labels1 = np.zeros((500, 1))  # Labels for dataset 1
labels2 = np.ones((500, 1))   # Labels for dataset 2
combined_labels = np.vstack((labels1, labels2))

# Creating final dataset with labels
final_data = np.hstack((combined_data, combined_labels))  # Shape (1000, 3)

# Displaying first few rows
print("First 5 rows of the labeled dataset:")
print(final_data[:5])
```
