```python
import matplotlib.pyplot as plt
import numpy as np

def compute_sorori_shinzaemon(n_days=100):
    """Calculate the number of rice grains in the Sorori Shinzaemon anecdote for each day.
    
    Parameters
    ----------
    n_days: int
        Number of days to receive rice (default: 100)
    
    Returns
    -------
    list_n_grains : list
        List of rice grains received on each day (length equal to n_days)
    list_total_grains : list
        List of cumulative rice grains received up to each day (length equal to n_days)
    """
    # Initialize lists to store daily and cumulative grains
    list_n_grains = []
    list_total_grains = []
    
    # Calculate rice grains for each day
    for day in range(1, n_days + 1):
        # On day n, receive 2^(n-1) grains
        grains_today = 2 ** (day - 1)
        list_n_grains.append(grains_today)
        
        # Calculate cumulative grains up to this day
        if day == 1:
            list_total_grains.append(grains_today)
        else:
            list_total_grains.append(list_total_grains[-1] + grains_today)
    
    return list_n_grains, list_total_grains

def plot_rice_grains(list_n_grains, list_total_grains, n_days):
    """Create a line graph showing daily and cumulative rice grains.
    
    Parameters
    ----------
    list_n_grains: list
        List of rice grains received on each day
    list_total_grains: list
        List of cumulative rice grains received up to each day
    n_days: int
        Number of days to display in the graph
    """
    days = list(range(1, n_days + 1))
    
    plt.figure(figsize=(12, 6))
    plt.plot(days, list_n_grains, label='Grains received on that day')
    plt.plot(days, list_total_grains, label='Cumulative grains')
    plt.xlabel('Days')
    plt.ylabel('Number of rice grains')
    plt.title(f'Rice Grains in Sorori Shinzaemon Problem over {n_days} Days')
    plt.grid(True)
    plt.legend()
    plt.yscale('log')  # Using log scale due to exponential growth
    plt.savefig(f'sorori_shinzaemon_{n_days}_days.png')
    plt.show()

def calculate_survival_days(total_grains, num_people):
    """Calculate how many days people can survive on a given amount of rice.
    
    Parameters
    ----------
    total_grains: int
        Total number of rice grains available
    num_people: int
        Number of people who need to eat
    
    Returns
    -------
    days: int
        Number of days people can survive
    
    Note: According to research, one person needs approximately 400 grams of rice per day
    for adequate nutrition. One grain of rice weighs about 0.029 grams on average.
    Therefore, one person needs about 13,793 grains of rice per day.
    Source: Various nutritional guidelines and agricultural data on rice grain weight.
    """
    # Research-based values
    grains_per_person_per_day = 13793  # approx. 400g of rice
    
    # Calculate total survival days
    total_person_days = total_grains / grains_per_person_per_day
    days = int(total_person_days / num_people)
    
    return days
```

```python
# [Question 1] Calculate rice grains for 100 days
n_days_q1 = 100
list_n_grains_q1, list_total_grains_q1 = compute_sorori_shinzaemon(n_days_q1)

print(f"[Question 1] Total rice grains after 100 days: {list_total_grains_q1[-1]}")
plot_rice_grains(list_n_grains_q1, list_total_grains_q1, n_days_q1)
```

```python
# [Question 2] Calculate rice grains for course duration (taking 15 weeks = 105 days)
course_days = 105  
list_n_grains_q2, list_total_grains_q2 = compute_sorori_shinzaemon(course_days)
print(f"\n[Question 2] Total rice grains after {course_days} days: {list_total_grains_q2[-1]}")
plot_rice_grains(list_n_grains_q2, list_total_grains_q2, course_days)
```

```python
# [Question 3] Calculate survival days for class
class_size = 30  
total_grains = list_total_grains_q2[-1]
survival_days = calculate_survival_days(total_grains, class_size)

print(f"\n[Question 3] A class of {class_size} people can survive for {survival_days} days")
print(f"on {total_grains} grains of rice received over {course_days} days.")
```
